```python
# connect to google 

from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360) 
```

```python
# build payload

kw_list = ["electric car", "ev", "tesla","electric vehicle", "hybrid"] # list of keywords to get data 

pytrends.build_payload(kw_list, cat=0, geo='US', timeframe='all') 
```

```python
#1 Interest over Time
data = pytrends.interest_over_time() 
data = data.reset_index() 


import plotly.express as px

fig = px.line(data, x="date", y=["electric car", "ev", "tesla","electric vehicle", "hybrid"], title='Keyword Web Search Interest Over Time')
fig.show() 
```

```python
by_region = pytrends.interest_by_region(resolution='REGION', inc_low_vol=True, inc_geo_code=False)

by_region.head(50) 
```
