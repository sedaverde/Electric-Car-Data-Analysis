```python
import pandas as pd
```

```python
file_path = (r"C:\Users\adaiu\OneDrive\Desktop\Analysis Projects\Electric-Car-Data-Analysis\Resources\EVs\EvsLightDutyDetails.csv")
light_duty_details_df = pd.read_csv(file_path)
light_duty_details_df.head()
```

```python
light_duty_details_df.info()
```

```python
light_duty_details_df.columns
```

```python
light_duty_details_df = light_duty_details_df[['Model Year','Model','Manufacturer','Manufacturer ID','Vehicle ID','Category',
                                               'Fuel','Fuel ID','Transmission Type', 'Engine Description',]]

light_duty_details_df.head()
```

```python
light_duty_details_df.drop(light_duty_details_df.loc[light_duty_details_df["Model Year"] < 2010].index, inplace = True)
light_duty_details_df.head()
```

```python
light_duty_details_df.dtypes
```

```python
light_duty_details_df.rename(columns = {'Model Year':'Model_Year', 'Manufacturer ID':'ManufacturerID',
                                        'Vehicle ID':'VehicleID', 'Fuel ID':'FuelID', 'Transmission Type':'Transmission_Type',
                                        'Engine Description':'Engine_Description'}, inplace = True)
light_duty_details_df.head()
```

```python
light_duty_details_df.sort_values(by=['Model_Year', 'Model', 'Manufacturer'], inplace=True)
light_duty_details_df.reset_index(inplace=True, drop= True)
light_duty_details_df.head()
```

```python
filepath = (r"C:\Users\adaiu\OneDrive\Desktop\Analysis Projects\Electric-Car-Data-Analysis\Resources\EVs\Cleaned_datasets\EvsLightDutyDetails_Cleaned.csv")
light_duty_details_df.to_csv(filepath, index = False)
```
