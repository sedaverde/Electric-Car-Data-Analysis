```python
import pandas as pd
from pathlib import Path
```

```python
file_path = (r"C:\Users\adaiu\OneDrive\Desktop\Analysis Projects\Electric-Car-Data-Analysis\Resources\EVs\Electric_Vehicle_Population_Size_History.csv")
ev_pop_size_df = pd.read_csv(file_path)
ev_pop_size_df.head()
```

```python
ev_pop_size_df["Date"] = pd.to_datetime(ev_pop_size_df["Date"], yearfirst=True)
ev_pop_size_df.head()
```

```python
ev_pop_size_df.sort_values('Date', inplace = True)
ev_pop_size_df.head()
```

```python
ev_pop_size_df.dtypes
```

```python
ev_pop_size_df.rename(columns = {'Plug-In Hybrid Electric Vehicle (PHEV) Count':'Plug-In_Hybrid_(PHEV)_Count',
                       'Battery Electric Vehicle (BEV) Count':'Battery_Electric_Vehicle_(BEV)_Count',
                      'Electric Vehicle (EV) Total':'Electric_Vehicle_(EV)_Total'}, inplace=True)
ev_pop_size_df.head()
```

```python
ev_pop_size_df['yyyy'] = pd.to_datetime(ev_pop_size_df['Date']).dt.year
ev_pop_size_df['mm'] = pd.to_datetime(ev_pop_size_df['Date']).dt.month
ev_pop_size_df['dateValue'] = ev_pop_size_df['yyyy'].astype(str) + '-' + ev_pop_size_df['mm'].astype(str)

ev_pop_size_df.head()
```

```python
ev_pop_size_df.drop(columns = ['Date', 'yyyy', 'mm'], inplace=True)
ev_pop_size_df.head()
```

```python
filepath = (r"C:\Users\adaiu\OneDrive\Desktop\Analysis Projects\Electric-Car-Data-Analysis\Resources\EVs\Cleaned_datasets\Electric_Vehicle_Population_Size_History_Cleaned.csv")
ev_pop_size_df.to_csv(filepath, index=False)
```
