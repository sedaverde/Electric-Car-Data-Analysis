```python
import pandas as pd
```

```python
file_path = (r"C:\Users\adaiu\OneDrive\Desktop\Analysis Projects\Electric-Car-Data-Analysis\Resources\Fuel Cost\gasolinePriceData.csv")
gasprice_df = pd.read_csv(file_path)
gasprice_df.head()
```

```python
gasprice_df.info()
```

```python
gasprice_df.drop(columns = ["Unnamed: 3", "Unnamed: 4","Unnamed: 5", "Unnamed: 6"], inplace = True)
```

```python
gasprice_df['date'] = pd.to_datetime(gasprice_df['date'], yearfirst=True)
gasprice_df.head()
```

```python
gasprice_df.drop(gasprice_df.loc[gasprice_df["date"] < "2010"].index, inplace = True)
gasprice_df.reset_index(drop=True, inplace = True)
gasprice_df
```

```python
gasprice_df.dtypes
```

```python
gasprice_df["dateId"] = gasprice_df["dateId"].str.replace(",","").astype(int)
gasprice_df.dtypes
```

```python
gasprice_df.head()
```

```python
gasprice_df['yyyy'] = pd.to_datetime(gasprice_df['date']).dt.year
gasprice_df['mm'] = pd.to_datetime(gasprice_df['date']).dt.month
gasprice_df['dateValue'] = gasprice_df['yyyy'].astype(str) + '-' + gasprice_df['mm'].astype(str)

gasprice_df.head()
```

```python
gasprice_df.drop(columns = ['date', 'yyyy', 'mm'], inplace=True)
gasprice_df.head()
```

```python
filepath = (r"C:\Users\adaiu\OneDrive\Desktop\Analysis Projects\Electric-Car-Data-Analysis\Resources\Fuel Cost\Cleaned_dataset\gasolinePriceData_Cleaned.csv")
gasprice_df.to_csv(filepath, index = False)
```
